```python
import os
import numpy as np
import pandas as pd

sessions = np.arange(25, 76)
data=[]

for session in sessions:
    directory = "./TXT/Session " + str(session) + " - " + str(1945 + session)
    for filename in os.listdir(directory):
        f = open(os.path.join(directory, filename), encoding='utf8')
        if filename[0] == ".":
            continue
        splt = filename.split("_")
        data.append([session, 1945 + session, splt[0], f.read()])

df_speech = pd.DataFrame(data, columns=['Session','Year','ISO-alpha3 Code','Speech'])
df_speech['Year'].value_counts()
```

```python
import matplotlib.pyplot as plt


speeches_per_year = df_speech['Year'].value_counts()
f, ax = plt.subplots(figsize=(8,5)) # set the size that you'd like (width, height)
plt.bar(speeches_per_year.index, speeches_per_year.values)
ax.set_ylabel('Number of speeches')
ax.set_xlabel('Year')
ax.set_title('Change in number of speeches over 50 years')
plt.show()
f.savefig('speeches.png')
```
